## DIA 107: Exportar Gráficas del Dashboard como Imagen o PDF (Chart.js + html2canvas + jsPDF)

Hoy aprenderás a permitir que el usuario descargue la gráfica de su dashboard como imagen PNG o PDF, útil para informes y presentaciones.

✅ ¿Qué aprenderás hoy?

🖼️ Capturar gráficos en HTML como imágenes.

📄 Generar un PDF con esa imagen.

🧲 Descargar ambos formatos al instante.

🧩 1. Librerías necesarias en el HTML
Agrega esto en tu <head> en dashboard.html:

<script>
  // (ya tienes código del gráfico aquí...)

  async function descargarPNG() {
    const canvas = document.getElementById('grafico');
    const link = document.createElement('a');
    link.download = 'grafico_predicciones.png';
    link.href = canvas.toDataURL('image/png');
    link.click();
  }

  async function descargarPDF() {
    const canvas = document.getElementById('grafico');
    
    // Usar html2canvas
    html2canvas(canvas).then(canvasCapturado => {
      const imgData = canvasCapturado.toDataURL('image/png');
      const { jsPDF } = window.jspdf;
      const pdf = new jsPDF();

      const imgWidth = 180; // mm
      const pageWidth = pdf.internal.pageSize.getWidth();
      const imgHeight = canvas.height * imgWidth / canvas.width;

      const x = (pageWidth - imgWidth) / 2;
      pdf.text("📊 Gráfico de Predicciones", 15, 15);
      pdf.addImage(imgData, 'PNG', x, 25, imgWidth, imgHeight);
      pdf.save("grafico_predicciones.pdf");
    });
  }
</script>

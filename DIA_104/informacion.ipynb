## DIA 104: Desplegar tu App Flask + API + Frontend en la Nube (Render)

¡Bienvenido al Día 104!
Hoy llevarás tu proyecto al siguiente nivel: lo subirás a internet usando Render.com, una plataforma fácil y gratuita (para pruebas) que permite desplegar apps Flask con base de datos y API.

✅ ¿Qué aprenderás hoy?

📤 Cómo preparar Flask para producción.

🌍 Subir tu app a Render.

🔗 Usar la API online desde cualquier lugar.

🧩 7. Desplegar en Render
Ve a https://render.com

Crea una cuenta y haz click en "New Web Service"

Conecta tu GitHub y elige el repo

Render detectará requirements.txt y .render.yaml

Dale un nombre, espera que se construya y ¡listo!

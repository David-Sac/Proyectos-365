## DIA 092: Mostrar Gráfica de Predicciones Solo del Usuario Logueado (Chart.js + Flask)

En días anteriores creamos gráficas globales con Chart.js.
Hoy aprenderás a crear una gráfica personalizada que muestre solo las predicciones realizadas por el usuario logueado.

✅ ¿Qué aprenderás hoy?

📌 Filtrar los datos por usuario usando current_user.

📌 Contar las predicciones por clase.

📌 Visualizarlas con Chart.js en una gráfica de barras.

🧩 1. Ruta Nueva en app.py → /mi-grafica

from collections import Counter

@app.route('/mi-grafica')
@login_required
def mi_grafica():
    registros = Registro.query.filter_by(user_id=current_user.id).all()
    predicciones = [r.prediccion for r in registros]

    conteo = Counter(predicciones)
    clases = list(conteo.keys())
    cantidades = list(conteo.values())

    return render_template('mi_grafica.html', clases=clases, cantidades=cantidades)

🔍 Líneas Clave Explicadas
Registro.query.filter_by(user_id=current_user.id): asegura que los datos son del usuario actual.

Counter(predicciones): cuenta cuántas veces se repite cada clase.

{{ clases|tojson }}: convierte listas Python a JSON para usar en JS.

from collections import Counter

@app.route('/mi-grafica')
@login_required
def mi_grafica():
    registros = Registro.query.filter_by(user_id=current_user.id).all()
    predicciones = [r.prediccion for r in registros]

    conteo = Counter(predicciones)
    clases = list(conteo.keys())
    cantidades = list(conteo.values())

    return render_template('mi_grafica.html', clases=clases, cantidades=cantidades)

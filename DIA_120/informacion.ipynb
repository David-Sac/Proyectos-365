## DIA 120: Visor Online de Reportes PDF y CSV sin Necesidad de Descargarlos

Hoy vas a agregar un visor integrado que permita al usuario ver sus reportes directamente en el navegador, sin tener que descargarlos manualmente.

✅ ¿Qué aprenderás hoy?

📄 Incrustar archivos PDF y CSV en HTML

🌐 Usar <iframe> y <object> para visores

🔐 Proteger la vista para que solo el usuario acceda a sus archivos

🧩 1. Crear ruta Flask para ver un reporte

@app.route('/ver-reporte/<int:reporte_id>')
@login_required
def ver_reporte(reporte_id):
    reporte = Reporte.query.get_or_404(reporte_id)

    # Verifica que el usuario tenga acceso
    if reporte.user_id != current_user.id:
        return "No tienes permiso para ver este reporte.", 403

    return render_template("ver_reporte.html", reporte=reporte)

  {% if reporte.tipo == 'pdf' %}
    <object data="{{ reporte.url }}" type="application/pdf">
      <p>Tu navegador no soporta PDFs. <a href="{{ reporte.url }}">Descargar PDF</a>.</p>
    </object>
  {% elif reporte.tipo == 'csv' %}
    <iframe src="https://docs.google.com/gview?url={{ reporte.url }}&embedded=true"></iframe>
    <p>Si no carga, <a href="{{ reporte.url }}" target="_blank">descargar CSV</a>.</p>
  {% else %}
    <p>Tipo de archivo no soportado para vista previa.</p>
  {% endif %}

from flask import request
from datetime import datetime

@app.route('/historial-reportes')
@login_required
def historial_reportes():
    tipo = request.args.get("tipo")
    fecha_inicio = request.args.get("inicio")
    fecha_fin = request.args.get("fin")

    query = Reporte.query.filter_by(user_id=current_user.id)

    if tipo:
        query = query.filter(Reporte.tipo == tipo)

    if fecha_inicio:
        fecha_ini_dt = datetime.strptime(fecha_inicio, "%Y-%m-%d")
        query = query.filter(Reporte.fecha_generado >= fecha_ini_dt)

    if fecha_fin:
        fecha_fin_dt = datetime.strptime(fecha_fin, "%Y-%m-%d")
        query = query.filter(Reporte.fecha_generado <= fecha_fin_dt)

    reportes = query.order_by(Reporte.fecha_generado.desc()).all()

    return render_template("historial_reportes.html", reportes=reportes, tipo=tipo,
                           fecha_inicio=fecha_inicio, fecha_fin=fecha_fin)

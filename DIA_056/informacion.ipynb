import os
import logging
from flask import Flask, render_template
from flask_socketio import SocketIO, send

# Configuraci√≥n b√°sica de Flask
app = Flask(__name__)
app.config['SECRET_KEY'] = os.getenv('SECRET_KEY', 'supersecretkey')

# Inicializar SocketIO
socketio = SocketIO(app, cors_allowed_origins="*")

# Configuraci√≥n de Logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# ---------------------------
# P√°gina Principal del Chat
# ---------------------------
@app.route('/')
def index():
    """Renderiza la p√°gina del chat."""
    return render_template("chat.html")

# ---------------------------
# Manejador de Mensajes WebSocket
# ---------------------------
@socketio.on('message')
def handle_message(msg):
    """Recibe mensajes del cliente y los retransmite a todos."""
    logger.info(f"üì© Mensaje recibido: {msg}")
    send(msg, broadcast=True)  # Reenviar a todos los clientes conectados

# ---------------------------
# Ejecutar la aplicaci√≥n
# ---------------------------
if __name__ == '__main__':
    socketio.run(app, debug=True)

from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer

# ---------------------------
# 1. Configuración del Chatbot
# ---------------------------
chatbot = ChatBot(
    "AsistenteTech",
    storage_adapter="chatterbot.storage.SQLStorageAdapter",
    database_uri="sqlite:///chatbot.db",
)

# ---------------------------
# 2. Entrenar el Chatbot
# ---------------------------
trainer = ChatterBotCorpusTrainer(chatbot)
trainer.train("chatterbot.corpus.spanish")

# ---------------------------
# 3. Función para Conversar con el Bot
# ---------------------------
def chat():
    print("🗣️ Chatbot AsistenteTech - Escribe 'salir' para finalizar.")
    
    while True:
        user_input = input("Tú: ")
        if user_input.lower() == "salir":
            print("👋 Adiós, ¡hasta pronto!")
            break
        response = chatbot.get_response(user_input)
        print(f"AsistenteTech: {response}")

# ---------------------------
# 4. Ejecutar el Chatbot
# ---------------------------
if __name__ == "__main__":
    chat()

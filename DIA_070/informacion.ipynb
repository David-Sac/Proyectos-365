from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, CallbackContext

# 1. Configurar el Token del Bot
# ---------------------------

# ---------------------------
# 2. Definir Comandos del Bot
# ---------------------------
def start(update: Update, context: CallbackContext):
    update.message.reply_text("ğŸ‘‹ Â¡Hola! Soy un bot de Telegram.\nUsa /help para ver los comandos disponibles.")

def help_command(update: Update, context: CallbackContext):
    update.message.reply_text("ğŸ“Œ Comandos disponibles:\n/start - Iniciar bot\n/help - Ver ayuda")

def echo(update: Update, context: CallbackContext):
    update.message.reply_text(f"ğŸ”„ Repetiste: {update.message.text}")

# ---------------------------
# 3. Configurar el Bot
# ---------------------------
def main():
    updater = Updater(TOKEN, use_context=True)
    dp = updater.dispatcher

    # Agregar comandos

    updater.idle()

if __name__ == "__main__":

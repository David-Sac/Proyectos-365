from telegram import Update
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, CallbackContext

# ---------------------------
# 1. Configurar el Token del Bot
# ---------------------------
TOKEN = "TU_TOKEN_AQUI"

# ---------------------------
# 2. Definir Comandos del Bot
# ---------------------------
def start(update: Update, context: CallbackContext):
    update.message.reply_text("👋 ¡Hola! Soy un bot de Telegram.\nUsa /help para ver los comandos disponibles.")

def help_command(update: Update, context: CallbackContext):
    update.message.reply_text("📌 Comandos disponibles:\n/start - Iniciar bot\n/help - Ver ayuda")

def echo(update: Update, context: CallbackContext):
    update.message.reply_text(f"🔄 Repetiste: {update.message.text}")

# ---------------------------
# 3. Configurar el Bot
# ---------------------------
def main():
    updater = Updater(TOKEN, use_context=True)
    dp = updater.dispatcher

    # Agregar comandos
    dp.add_handler(CommandHandler("start", start))
    dp.add_handler(CommandHandler("help", help_command))
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, echo))

    # Iniciar el bot
    updater.start_polling()
    updater.idle()

if __name__ == "__main__":
    main()

🔍 Explicación de las Principales Implementaciones
🔹 📂 Uso de Updater y Dispatcher

Updater.start_polling() mantiene el bot en ejecución.
Dispatcher gestiona los mensajes recibidos.
🔹 🌍 Manejo de Comandos con CommandHandler

/start y /help muestran mensajes de bienvenida y ayuda.
🔹 🔄 Respuesta Automática con MessageHandler

El bot repite cualquier mensaje recibido.

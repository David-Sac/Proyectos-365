## DIA 090: Mostrar el Historial de Predicciones por Usuario Logueado en Flask

Ahora que tienes un sistema de autenticación funcional, es momento de mostrar a cada usuario su propio historial de predicciones, es decir, solo podrá ver sus datos y no los de otros.

✅ ¿Qué aprenderás hoy?

📌 Consultar registros filtrados por el usuario logueado.

📌 Mostrar solo las predicciones que le pertenecen.

📌 Mejorar la privacidad y organización de los datos en tu app.

🧩 1. 🧠 Lógica en Flask (app.py)
Agrega esta nueva ruta protegida por @login_required:

python
Copiar
Editar
from flask_login import login_required, current_user

@app.route('/mi-historial')
@login_required
def mi_historial():
    registros = Registro.query.filter_by(user_id=current_user.id).order_by(Registro.fecha.desc()).all()
    return render_template("mi_historial.html", registros=registros)

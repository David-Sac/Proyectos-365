@app.route('/mi-grafica-fechas')
@login_required
def mi_grafica_fechas():
    registros = Registro.query.filter_by(user_id=current_user.id).all()

    fechas = [r.fecha.date() for r in registros]
    conteo = Counter(fechas)

    labels = [fecha.strftime('%Y-%m-%d') for fecha in sorted(conteo.keys())]
    valores = [conteo[fecha] for fecha in sorted(conteo.keys())]

    return render_template("mi_grafica_fechas.html", labels=labels, valores=valores)

import qrcode

# ---------------------------
# 1. Datos para el Código QR
# ---------------------------
data = "https://www.python.org"  # URL o texto a codificar

# ---------------------------
# 2. Configurar el Código QR
# ---------------------------
qr = qrcode.QRCode(
    version=5,  # Tamaño del código QR (1 a 40)
    error_correction=qrcode.constants.ERROR_CORRECT_L,  # Nivel de corrección de errores
    box_size=10,  # Tamaño de cada caja del QR
    border=4  # Grosor del borde blanco alrededor del QR
)

qr.add_data(data)  # Agregar los datos al QR
qr.make(fit=True)  # Ajustar el tamaño automáticamente

# ---------------------------
# 3. Generar el Código QR
# ---------------------------
img = qr.make_image(fill="black", back_color="white")  # Generar imagen QR

# ---------------------------
# 4. Guardar la Imagen QR
# ---------------------------
img.save("codigo_qr.png")

print("✅ Código QR generado y guardado como 'codigo_qr.png'")

🔍 Explicación de las Principales Implementaciones
🔹 📂 Configuración del Código QR

version=5 define el tamaño del QR (cuanto mayor, más datos puede almacenar).
error_correction=ERROR_CORRECT_L permite corregir errores leves en la lectura.
🔹 🔄 Personalización del Código QR

box_size=10 define el tamaño de los cuadrados en el QR.
fill="black", back_color="white" permite cambiar los colores.
🔹 🖼️ Guardado y Exportación

img.save("codigo_qr.png") guarda el QR como imagen PNG.

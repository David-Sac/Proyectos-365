## DIA 001: Exploracion avanzada de datos en Python

La exploración (o EDA: Exploratory Data Analysis) es el proceso inicial en la ciencia de datos y en cualquier proyecto analítico, donde se estudian y comprenden las características principales de los datos. Esto incluye identificar distribuciones, valores atípicos (outliers), correlaciones, patrones o tendencias. Generalmente, esta exploración se apoya en bibliotecas como Pandas (para manipulación de datos) y Matplotlib o Seaborn (para visualización).

```python
# 1.Cargar datos
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
a = "iris/iris.data"
db = pd.read_csv(a, header=None, names=column_names)
```

```python
# 2.Exploracion Inicial
print("Primeras filas - si no pones algo dentro de () se muestra automaticamente 5")
print(db.head())
print("===============================================================================")
print("Informacion General")
print(db.info())
print("Muestra el tamaño del archivo")
print(db.shape)
print("Describe el archivo")
print(db.describe())
```

```python
# 3.Limpieza básica
print("Encontraremos la cantidad de valores nulos por columna")
print(db.isnull().sum())
```

```python
# 4.Realizaremos visualizaciones exploratorias
# Verifica que los datos se cargaron correctamente
print(db.head())
print(db.columns)

# Realizar visualizaciones exploratorias
print("Distribución de cada variable numérica")
db.hist(figsize=(8, 6), bins=10)
plt.tight_layout()
plt.show()

print("Boxplot para detectar outliers")
sns.boxplot(data=db.drop(columns="class"))
plt.title("Boxplot de variables numéricas")
plt.show()

print("Countplot de la variable categórica")
sns.countplot(x="class", data=db)
plt.title("Distribución por especie")
plt.show()

print("Matriz de correlación")
corr = db.drop(columns="class").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap de correlación")
plt.show()
```

# ---------------------------

st.set_page_config(page_title="ðŸ“Š Dashboard de Ventas", layout="wide")

st.title("ðŸ“Š Dashboard de Ventas Interactivo")

# ---------------------------
# 2. Cargar Datos

@st.cache_data
def cargar_datos():
    data = {
        "Fecha": pd.date_range(start="2023-01-01", periods=100, freq="D"),
        "Ventas": [abs(int(x)) for x in (1000 + 500 * pd.np.random.randn(100))]
    }
    return pd.DataFrame(data)

df = cargar_datos()

# ---------------------------
# 3. SelecciÃ³n de Rango de Fechas

st.sidebar.header("ðŸ“… Filtros")

# ---------------------------
# 4. GrÃ¡fico de Ventas
# ---------------------------

# ---------------------------
# 5. Mostrar Datos Filtrados

import dash
from dash import dcc, html
import pandas as pd
import plotly.express as px
from dash.dependencies import Input, Output

# ---------------------------
# 1. Cargar Datos
# ---------------------------
df = pd.DataFrame({
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo"],
    "Ventas": [12000, 15000, 18000, 13000, 20000],
    "CategorÃ­a": ["ElectrÃ³nica", "Ropa", "ElectrÃ³nica", "Hogar", "Ropa"]
})

# ---------------------------
# 2. Inicializar Dash App
# ---------------------------
app = dash.Dash(__name__)

# ---------------------------
# 3. DiseÃ±o del Dashboard
# ---------------------------
app.layout = html.Div([
    html.H1("ğŸ“Š Dashboard de Ventas"),
    
    dcc.Dropdown(
        id="categoria-dropdown",
        options=[{"label": cat, "value": cat} for cat in df["CategorÃ­a"].unique()],
        value="ElectrÃ³nica",
        clearable=False
    ),
    
    dcc.Graph(id="ventas-grafico"),
    
    dcc.Slider(
        id="mes-slider",
        min=0,
        max=len(df["Mes"]) - 1,
        marks={i: mes for i, mes in enumerate(df["Mes"])},
        value=0,
        step=1
    )
])

# ---------------------------
# 4. Callbacks para Interactividad
# ---------------------------
@app.callback(
    Output("ventas-grafico", "figure"),
    [Input("categoria-dropdown", "value"),
     Input("mes-slider", "value")]
)
def actualizar_grafico(categoria, mes_index):
    df_filtrado = df[(df["CategorÃ­a"] == categoria) & (df.index == mes_index)]
    
    fig = px.bar(df_filtrado, x="Mes", y="Ventas", color="CategorÃ­a", title="Ventas por Mes")
    
    return fig

# ---------------------------
# 5. Ejecutar la AplicaciÃ³n
# ---------------------------
if __name__ == "__main__":
    app.run_server(debug=True)

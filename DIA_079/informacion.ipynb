import tkinter as tk

# ---------------------------
# 1. Crear la ventana principal
# ---------------------------
ventana = tk.Tk()
ventana.title("Calculadora")
ventana.geometry("300x400")
ventana.resizable(False, False)

# ---------------------------
# 2. Campo de entrada (pantalla)
# ---------------------------
entrada = tk.Entry(ventana, font=("Arial", 24), borderwidth=5, relief="sunken", justify="right")
entrada.grid(row=0, column=0, columnspan=4, padx=10, pady=10, ipadx=8, ipady=15)

# ---------------------------
# 3. Funciones de la calculadora
# ---------------------------
def click(boton):
    actual = entrada.get()
    entrada.delete(0, tk.END)
    entrada.insert(0, actual + str(boton))

def limpiar():
    entrada.delete(0, tk.END)

def calcular():
    try:
        resultado = eval(entrada.get())
        entrada.delete(0, tk.END)
        entrada.insert(0, str(resultado))
    except:
        entrada.delete(0, tk.END)
        entrada.insert(0, "Error")

# ---------------------------
# 4. Botones de la calculadora
# ---------------------------
botones = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('.', 4, 1), ('C', 4, 2), ('+', 4, 3),
]

for (texto, fila, columna) in botones:
    if texto == 'C':
        accion = limpiar
    else:
        accion = lambda x=texto: click(x)
    tk.Button(ventana, text=texto, width=5, height=2, font=("Arial", 16), command=accion)\
        .grid(row=fila, column=columna, padx=5, pady=5)

# Bot√≥n de igual (=)
tk.Button(ventana, text='=', width=20, height=2, font=("Arial", 16), bg="lightblue", command=calcular)\
    .grid(row=5, column=0, columnspan=4, padx=5, pady=10)

# ---------------------------
# 5. Ejecutar la interfaz
# ---------------------------
ventana.mainloop()
üîç Explicaci√≥n de las l√≠neas m√°s importantes
tk.Entry(...): Crea el campo donde se muestra la operaci√≥n y el resultado.
click(boton): Agrega el texto del bot√≥n presionado al campo de entrada.
eval(...): Eval√∫a la expresi√≥n matem√°tica ingresada como una operaci√≥n real.
tk.Button(...): Crea cada bot√≥n con su posici√≥n en la grilla.
ventana.mainloop(): Mantiene la ventana abierta esperando eventos del usuario.
